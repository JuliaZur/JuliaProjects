# ZADANIE 1

```julia
Fs = 1024
t = 0:1/(Fs-1):1
x = sin.(2*pi*t*200) + 2* sin.(2*pi*t*400) 
```

```julia
using Plots
```

```julia
plot(t,x)
```

```julia
using FFTW
y=fft(x)
```

```julia
sticks(abs.(y))
```

Według spodziewań najwyższa amplituda pojawia się dla częstotliwości pierwszego sygnału - 200 Hz
oraz dla częstotliwości drugiego sygnału - 400 Hz. A także dla ich symetrycznych odbić.

# ZADANIE 2

```julia
Fs = 1024
t = 0:1/(Fs-1):1
x2 = 2*cos.(2*pi*t*200) 
r = []
for i in 0:1/(Fs-1):1
    push!(r,rand(Float64))
end
x2 = r+x2
```

```julia
plot(t,x2)
```

```julia
y2 = fft(x2)
```

```julia
sticks(abs.(y2))
```

```julia
a = abs.(y2)

for i in 1:1:Fs
    if(abs.(a[i])) < 50
        a[i] = 0
    end
end
a
```

```julia
sticks(a)
```

```julia
b = ifft(a)
```

```julia
plot(t,real.(b), label="cleaned signal")
```

```julia
plot!(t,x2,label="input signal")
```

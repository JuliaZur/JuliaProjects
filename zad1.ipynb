```r
results = read.csv("results.csv")
```

```r
results
```

```r
avg_results_naive = aggregate(Naive ~Size, data=results, FUN=mean)
```

```r
avg_results_naive
```

```r
avg_results_blas = aggregate(Blas ~Size, data=results, FUN=mean)
```

```r
avg_results_blas
```

```r
avg_results_better = aggregate(Better ~Size, data=results, FUN=mean)
```

```r
avg_results_better
```

```r
STD_results_naive = aggregate(Naive ~Size, data=results, FUN=sd)
```

```r
STD_results_naive
```

```r
STD_results_blas = aggregate(Blas ~Size, data=results, FUN=sd)
```

```r
STD_results_blas
```

```r
STD_results_better = aggregate(Better ~Size, data=results, FUN=sd)
```

```r
STD_results_better
```

```r
naive_results = avg_results_naive
```

```r
naive_results
```

```r
naive_results$STD = STD_results_naive$Naive
```

```r
naive_results
```

```r
better_results = avg_results_better
```

```r
better_results$STD = STD_results_better$Better
```

```r
better_results
```

```r
blas_results = avg_results_blas
```

```r
blas_results$STD = STD_results_blas$Blas
```

```r
blas_results
```

```r
naive_plot = ggplot(naive_results, aes(Size,Naive)) + geom_point() + geom_line() + ggtitle("Naive") + ylab("Time [ns]") + geom_errorbar(aes(ymin=Naive-STD/2,ymax=Naive+STD/2),colour="blue")
```

```r
naive_plot
```

```r
better_plot = ggplot(better_results, aes(Size,Better)) + geom_point() + geom_line() + ggtitle("Better") + ylab("Time [ns]") + geom_errorbar(aes(ymin=Better-STD/2,ymax=Better+STD/2),colour="green")
better_plot
```

```r
blas_plot = ggplot(blas_results, aes(Size,Blas)) + geom_point() + geom_line() + ggtitle("Blas") + ylab("Time [ns]") + geom_errorbar(aes(ymin=Blas-STD/2,ymax=Blas+STD/2),colour="yellow")
blas_plot
```

```r
results_plot = ggplot() + geom_line(data = naive_results, mapping = aes(Size,Naive,colour = "Naive")) + geom_line(data = better_results, mapping = aes(Size,Better,colour = "Better")) + geom_line(data = blas_results, mapping = aes(Size,Blas,colour="Blas")) + ylab("Time [ns]")+ggtitle("Blas vs Naive vs Better")
results_plot
```

```r
Size = naive_results$Size
Time = naive_results$Naive
STD = naive_results$STD
mydata = data.frame(Size,Time)

fit = lm(Time ~ poly(Size, 3, raw=TRUE), data=mydata)
fit

plot = ggplot(mydata,aes(Size,Time)) + geom_point()
newdata = data.frame(Size = seq(100,1000, length.out=250))
newdata$Time = predict(fit, newdata)
last_plot() + geom_line(data=newdata, aes(Size,Time)) + geom_errorbar(aes(ymin=Time-STD/2,ymax=Time+STD/2),colour="yellow")
```

```r
Size = better_results$Size
Time = better_results$Better
STD = better_results$STD
mydata = data.frame(Size,Time)

fit = lm(Time ~ poly(Size, 3, raw=TRUE), data=mydata)
fit

plot = ggplot(mydata,aes(Size,Time)) + geom_point()
newdata = data.frame(Size = seq(100,1000, length.out=250))
newdata$Time = predict(fit, newdata)
last_plot() + geom_line(data=newdata, aes(Size,Time)) + geom_errorbar(aes(ymin=Time-STD/2,ymax=Time+STD/2),colour="pink")
```

```r
Size = blas_results$Size
Time = blas_results$Blas
STD = blas_results$STD
mydata = data.frame(Size,Time)

fit = lm(Time ~ poly(Size, 3, raw=TRUE), data=mydata)
fit

plot = ggplot(mydata,aes(Size,Time)) + geom_point()
newdata = data.frame(Size = seq(100,1000, length.out=250))
newdata$Time = predict(fit, newdata)
last_plot() + geom_line(data=newdata, aes(Size,Time)) + geom_errorbar(aes(ymin=Time-STD/2,ymax=Time+STD/2),colour="blue")
```

```r
fit_naive = lm(Naive ~ poly(Size, 3, raw=TRUE), data=naive_results)
fit_better = lm(Better ~ poly(Size, 3, raw=TRUE), data=better_results)
fit_blas = lm(Blas ~ poly(Size, 3, raw=TRUE), data=blas_results)

newdata_naive = data.frame(Size = seq(100,1000, length.out=250))
newdata_better = data.frame(Size = seq(100,1000, length.out=250))
newdata_blas = data.frame(Size = seq(100,1000, length.out=250))

newdata_naive$Time = predict(fit_naive, newdata_naive)
newdata_better$Time = predict(fit_better, newdata_better)
newdata_blas$Time = predict(fit_blas, newdata_blas)

results_plot = ggplot() + geom_point(data = naive_results, mapping = aes(Size,Naive,colour = "Naive")) + geom_point(data = better_results, mapping = aes(Size,Better,colour = "Better")) + geom_point(data = blas_results, mapping = aes(Size,Blas,colour="Blas")) + ylab("Time [ns]")+ggtitle("Blas vs Naive vs Better")
last_plot() + geom_line(data=newdata_naive, aes(Size,Time,colour="Naive")) + geom_line(data=newdata_better, aes(Size,Time,colour="Better")) + geom_line(data=newdata_blas, aes(Size,Time,colour="Blas"))
```
